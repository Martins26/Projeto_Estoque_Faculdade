```python
estoque = {
    1: {"categoria": "fruta", "preco": 19.0, "quantidade": 3},
    2: {"categoria": "verdura", "preco": 2.0, "quantidade": 5}
}

def cadastrar_produtos(nome, categoria, preco, quantidade):
  produtos = {
      "categoria": categoria,
      "preco": preco,
      "quantidade": quantidade
  }

  estoque[nome] = produtos
  print(f"Produto {nome} da categoria {categoria} cadastrado com sucesso!")


```

```python
def registrar_entrada(produto_id, quantidade):
    if produto_id in estoque:
      estoque[produto_id]["quantidade"] += quantidade
      print(f"Adicionado {quantidade} {produto_id}s!")
    else:
      print("Produto não encontrado")
```

```python
def registrar_saida(produto_id, quantidade):
    if produto_id in estoque:
        if estoque[produto_id]["quantidade"] >= quantidade:
            estoque[produto_id]["quantidade"] -= quantidade
            print(f"Removido {quantidade} {produto_id}s!")
        else:
            print("Quantidade indisponivel")
    else:
        print("Produto não encontrado")
```
