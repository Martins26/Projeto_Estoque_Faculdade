```python
def cadastrar_produto(estoque, produto, quantidade, preco):

  if produto in estoque:
    estoque[produto][0] += quantidade
    estoque[produto][1] = preco
    print(f"Produto {produto} atualizado. Nova quantidade: {estoque[produto][0]}, Novo preço: R${estoque[produto][1]:.2f}")
  else:
      estoque[produto] = [quantidade, preco]
      print(f"Produto {produto} adicionado ao estoque. Quantidade: {quantidade}, Preço: R${preco:}")

meu_estoque = {}

cadastrar_produto(meu_estoque, "Café", 100, 15.50)
cadastrar_produto(meu_estoque, "Leite", 50, 4.20)

cadastrar_produto(meu_estoque, "Café", 20, 16.00)

cadastrar_produto(meu_estoque,"Pão", 30,8.90)

print("\nEstoque atual:")
for produto, dados in meu_estoque.items():
  print(f"-{produto} {dados[0]} {dados[1]}")
```
